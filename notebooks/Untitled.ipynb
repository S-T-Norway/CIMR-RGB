```python
import matplotlib.pyplot as plt
from shapely.geometry import Polygon
import geopandas as gpd
from matplotlib.patches import Polygon as MplPolygon

# Example `geospatial_bounds` in WKT format or as a bounding box (adjust based on your dataset)
geospatial_bounds = "POLYGON((-125.0 35.0, -125.0 45.0, -115.0 45.0, -115.0 35.0, -125.0 35.0))"

# Parse WKT Polygon using Shapely
polygon = Polygon([(float(coord.split()[0]), float(coord.split()[1])) 
                   for coord in geospatial_bounds.strip("POLYGON()").split(",")])

# Convert to a GeoDataFrame for better handling
gdf = gpd.GeoDataFrame([1], geometry=[polygon], crs="EPSG:4326")

# Plot using Matplotlib
fig, ax = plt.subplots(figsize=(10, 6))
gdf.boundary.plot(ax=ax, color="blue", linewidth=2, label="Geospatial Bounds")

# Add basemap (optional)
ax.set_title("Geospatial Bounds Visualization", fontsize=14)
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
plt.legend()
plt.grid(True)
plt.show()
```
